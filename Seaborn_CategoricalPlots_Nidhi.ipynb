```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
#to get estimators other than the mean, one can add the estimator agrument and pass it in the sns.barplot() where 
#x=the categorical variable, y=the numerical variable, data= the dataset and estimator=whichever statistic one wants.
```

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator= np.std )
```

```python
sns.countplot(x='sex', data=tips)
```

```python
sns.boxplot(x='day',y='total_bill', data=tips)
```

```python
sns.boxplot(x='day',y='total_bill', data=tips, hue='smoker')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex')
```

```python
#better way to compare it is to have the distributions side by side by using split=True argument
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue= 'sex', split=True)
```

```python
sns.stripplot(x='day', y= 'total_bill', data=tips)
```

```python
sns.stripplot(x='day', y= 'total_bill', data=tips, jitter=True)
```

```python
sns.stripplot(x='day', y= 'total_bill', data=tips, jitter=True, hue= 'sex')
```

```python
sns.stripplot(x='day', y= 'total_bill', data=tips, jitter=True, hue= 'sex', split=True)
```

```python
#swarmplots are like a combination of stripplots and violin plots
#swarmplots are not advised for very large datasets
```

```python
sns.swarmplot(x='day', y= 'total_bill', data=tips)
```

```python
#to combine swarmplots and violinplots we can use the following code
```

```python
sns.violinplot(x='day', y= 'total_bill', data=tips)
sns.swarmplot(x='day', y= 'total_bill', data=tips, color='black')
```

```python
#most generic way to create a plot is using sns.factorplot and then passing the kind argument within 
#it along with the x, y, and data
```

```python
sns.factorplot(x='day', y= 'total_bill', data=tips, kind='bar')
```
